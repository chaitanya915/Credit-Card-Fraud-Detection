Step 2: Handle Missing Values

```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH = "../data/raw/creditcard.csv"

df = pd.read_csv(DATA_PATH)

working_df = df.copy()
```

```python
working_df.isnull().sum()
```

```python
total_missing = working_df.isnull().sum().sum()

print("Total Missing Values:", total_missing)
```

```python
missing_percentage = (
    working_df.isnull().sum() /
    len(working_df)
) * 100

print(missing_percentage)
```

```python
missing_report = pd.DataFrame({
    "Missing Count": working_df.isnull().sum(),
    "Missing Percentage": missing_percentage
})

missing_report
```

```python
missing_columns = missing_report[
    missing_report["Missing Count"] > 0
]

missing_columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

sns.heatmap(
    working_df.isnull(),
    cbar=False,
    cmap="viridis"
)

plt.title("Missing Value Heatmap")

plt.show()
```

```python
missing_report.to_csv(
    "../reports/missing_value_report.csv",
    index=True
)
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    working_df.isnull(),
    cbar=False
)

plt.savefig(
    "../images/missing_values.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

Step 3: Detect and Remove Duplicate Transactions

```python
import pandas as pd
```

```python
DATA_PATH = "../data/raw/creditcard.csv"

df = pd.read_csv(DATA_PATH)

working_df = df.copy()
```

```python
print("Original Shape:", working_df.shape)
```

```python
duplicates = working_df.duplicated()

duplicates
```

```python
duplicate_count = working_df.duplicated().sum()

print("Duplicate Rows:", duplicate_count)
```

```python
duplicate_percentage = (
    duplicate_count /
    len(working_df)
) * 100

print(f"Duplicate Percentage: {duplicate_percentage:.3f}%")
```

```python
duplicate_rows = working_df[
    working_df.duplicated()
]

duplicate_rows.head()
```

```python
working_df[
    working_df.duplicated(keep=False)
].head(10)
```

```python
working_df = working_df.drop_duplicates()
```

```python
print("New Shape:", working_df.shape)
```

```python
working_df.duplicated().sum()
```

```python
comparison = pd.DataFrame({
    "Stage": ["Before Cleaning", "After Cleaning"],
    "Rows": [len(df), len(working_df)]
})

comparison
```

```python
working_df.to_csv(
    "../data/processed/creditcard_no_duplicates.csv",
    index=False
)
```

```python
duplicate_report = pd.DataFrame({
    "Original Rows": [len(df)],
    "Duplicate Rows": [duplicate_count],
    "Remaining Rows": [len(working_df)]
})

duplicate_report.to_csv(
    "../reports/duplicate_report.csv",
    index=False
)
```

```python
import matplotlib.pyplot as plt

labels = ["Before", "After"]
rows = [len(df), len(working_df)]

plt.figure(figsize=(6,5))

plt.bar(labels, rows)

plt.title("Dataset Size Before and After Duplicate Removal")
plt.ylabel("Number of Rows")

plt.show()
```

```python
plt.savefig(
    "../images/duplicate_removal.png",
    dpi=300,
    bbox_inches="tight"
)
```
